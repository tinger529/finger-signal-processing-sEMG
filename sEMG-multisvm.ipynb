```python
import numpy as np
import os
from multilabel import *
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

## Prepare Data

```python
random_seed = 20230531

files = [
    'rest.csv',
    '12.csv',
    '13.csv',
    '14.csv',
    '15.csv',
    '23.csv',
    '24.csv',
    '25.csv',
    '34.csv',
    '35.csv',
    '45.csv',
    '123.csv',
    '124.csv',
    '125.csv',
    '234.csv',
    '235.csv',
    '1234.csv',
    '1235.csv',
    '1245.csv',
    '1345.csv',
    '2345.csv',
    '12345.csv'
]

# Load data
path = os.path.abspath(os.getcwd()) # Get current path
X_train, labels_train, list_ch_train, X_test, labels_test, list_ch_test = read_training_data(
    data_path='%s%sdatasets/dataset4%s' % (path, os.sep, os.sep),
    files=files,
    shuffle=True,
    sub_split=True,
)

assert list_ch_train == list_ch_test, 'Mismatch in channels!'
```

```python
X_train, X_test = standardize(X_train, X_test)
```

```python
print(X_train[2200], X_train.shape)
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1) 
```

## Fit and Evaluate Model

```python
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score

# 建立SVM分類器
svm_classifier = SVC()

# 將SVM分類器轉換為多標籤分類器
multi_label_classifier = MultiOutputClassifier(svm_classifier)

# 訓練多標籤分類器
multi_label_classifier.fit(X_train, labels_train)

# 預測新的樣本
y_pred = multi_label_classifier.predict(X_test)
accuracy = accuracy_score(labels_test, y_pred)
print("Accuracy:", accuracy)
```

```python
l = len(labels_test)
score = 0
for i in range(l):
    for j in range(5):
        if labels_test[i][j] == y_pred[i][j]:
            score += 1
print('score:', score / (l * 5))
```
